```python
# ========================
# 1. Arithmetic Operators
# ========================

a = 10
b = 3

print("Arithmetic Operators:")
print("a + b =", a + b)     # Addition
print("a - b =", a - b)     # Subtraction
print("a * b =", a * b)     # Multiplication
print("a / b =", a / b)     # Division (float)
print("a // b =", a // b)   # Floor Division
print("a % b =", a % b)     # Modulus
print("a ** b =", a ** b)   # Exponentiation
print("-" * 30)

# ========================
# 2. Assignment Operators
# ========================

x = 5
print("Assignment Operators:")
x += 2
print("x += 2 →", x)
x -= 1
print("x -= 1 →", x)
x *= 3
print("x *= 3 →", x)
x /= 2
print("x /= 2 →", x)
x //= 2
print("x //= 2 →", x)
x %= 2
print("x %= 2 →", x)
x = 4
x **= 2
print("x **= 2 →", x)
x = 10
x &= 4
print("x &= 4 →", x)
x |= 2
print("x |= 2 →", x)
x ^= 5
print("x ^= 5 →", x)
x >>= 1
print("x >>= 1 →", x)
x <<= 2
print("x <<= 2 →", x)
print("-" * 30)

# ============================
# 3. Comparison Operators
# ============================

a = 7
b = 5
print("Comparison Operators:")
print("a == b:", a == b)
print("a != b:", a != b)
print("a > b:", a > b)
print("a < b:", a < b)
print("a >= b:", a >= b)
print("a <= b:", a <= b)
print("-" * 30)

# ========================
# 4. Logical Operators
# ========================

x = True
y = False

print("Logical Operators:")
print("x and y:", x and y)
print("x or y:", x or y)
print("not x:", not x)
print("-" * 30)

# ========================
# 5. Identity Operators
# ========================

a = [1, 2]
b = [1, 2]
c = a

print("Identity Operators:")
print("a is c:", a is c)
print("a is b:", a is b)
print("a is not b:", a is not b)
print("-" * 30)

# =========================
# 6. Membership Operators
# =========================

nums = [10, 20, 30]

print("Membership Operators:")
print("20 in nums:", 20 in nums)
print("50 not in nums:", 50 not in nums)
print("-" * 30)

# =======================
# 7. Bitwise Operators
# =======================

a = 5   # 0101
b = 3   # 0011

print("Bitwise Operators:")
print("a & b =", a & b)     # AND
print("a | b =", a | b)     # OR
print("a ^ b =", a ^ b)     # XOR
print("~a =", ~a)           # NOT (inverts bits)
print("a << 1 =", a << 1)   # Left Shift
print("a >> 1 =", a >> 1)   # Right Shift
print("-" * 30)
```
